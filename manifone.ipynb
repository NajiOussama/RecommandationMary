```python


# Chemin vers le dossier contenant les fichiers CSV
dossier_csv = '/Users/oussamanaji/Downloads/csvs'

# Liste pour stocker chaque DataFrame
dataframes = []

# Parcours de tous les fichiers dans le dossier
for fichier in os.listdir(dossier_csv):
    chemin_complet = os.path.join(dossier_csv, fichier)
    df = pd.read_csv(chemin_complet,encoding='latin1', delimiter=';')
    dataframes.append(df)

# Combiner tous les DataFrames en un seul
df_final = pd.concat(dataframes, ignore_index=True)

# Afficher les premières lignes du DataFrame final
df_final.head()
```

```python
df_final['CLI envoyé'] = df_final['CLI envoyé'].astype(str)
df_final['CLI envoyé'].head()
```

```python
numero_vw = ['33162563937', '33270490602', '33377480618', '33424500649', '33568500594']
df_final_vw = df_final[df_final['CLI envoyé'].isin(numero_vw)]
```
